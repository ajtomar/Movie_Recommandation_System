```python
import numpy as np
import pandas as pd
```

```python
df1= pd.read_csv("tmdb_5000_credits.csv")
```

```python
df2=pd.read_csv("tmdb_5000_movies.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.columns = ['id','tittle','cast','crew']
df2= df2.merge(df1,on='id')
```

```python
df2.head()
```

```python
# calculating the mean rating

c=df2['vote_average'].mean()
```

```python
c
```

```python
# The next step is to determine an appropriate value for m, the minimum votes required to be listed in the chart.
# We will use 90th percentile as our cutoff. In other words, for a movie to feature in the charts, it must have
# more votes than at least 90% of the movies in the list.

m= df2['vote_count'].quantile(0.9)
```

```python
m
```

```python
# Now, we can filter out the movies that qualify for the chart

q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
# movies which have qualified to be present in the list:

q_movies.head()
```

```python
# We see that there are 481 movies which qualify to be in this list. Now, we need to calculate our metric for each
# qualified movie. To do this, we will define a function, weighted_rating() and define a new feature score, of which
# we'll calculate the value by applying this function to our DataFrame of qualified movies:

def weighted_rating(x, m=m, C=c):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Define a new feature 'score' and calculate its value with `weighted_rating()`

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies.head()
```

```python
# Let's sort the DataFrame based on the score feature and output the title, vote count, vote average, popularity, and weighted rating or score of the top 10 movies.

q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'popularity', 'score']].head(10)
```

```python
pop= q_movies.sort_values('score', ascending=False)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,4))

plt.barh(pop['title'].head(6),pop['score'].head(6), align='center',
        color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular Movies")
```

```python
# Let's sort the DataFrame based on the popularity feature and output the title, vote count, vote average, score and populairty of the top 10 movies.


p_movies = q_movies.sort_values('popularity', ascending=False)
p_movies[['title', 'vote_count', 'vote_average','score', 'popularity']].head(10)
```

```python
pop= q_movies.sort_values('popularity', ascending=False)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,4))

plt.barh(pop['title'].head(6),pop['popularity'].head(6), align='center',
        color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular Movies")
```
